```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

# Read the CSV file containing URLs
df = pd.read_csv(r"D:\raptor supplies\19dec\skygeek\skygeek_links_product.csv", header=None)
df.columns = ['url']

# Initialize an empty list to store data
data_list = []

# Loop through each URL in the DataFrame
for url in df['url'].iloc[0:]:
    # Send a request to the URL
    data = requests.get(url).text
    soup = BeautifulSoup(data, "html.parser")

    # Extract image source
    imageg = soup.find('div', class_='productView-img-container')
    image_tag = imageg.find('a', itemprop='image') if imageg else None
    image_src = image_tag.find('img')['src'] if image_tag else None

    # Extract title
    title = soup.find('h1', class_='productView-title').text if soup.find('h1', class_='productView-title') else None

    # Extract manufacturer part number
    mpn_div1 = soup.find('div', class_='productView product-view-configurable')
    mpn_div = mpn_div1.find('div', class_='mpn-value italic-content') if mpn_div1 else None
    mfg_span = mpn_div.find('span').find_next_sibling('span') if mpn_div else None
    mfg = mfg_span.text.strip() if mfg_span else None

    # Extract price
    price_span = soup.find('div', class_='price-section price-section--withoutTax')
    price_ = price_span.find('span', class_='price price--withoutTax') if price_span else None
    price = price_.text.strip() if price_ else None

    # Extract breadcrumbs
    breadcrumbs_ul = soup.find('ul', class_='breadcrumbs')
    breadcrumbs_li_list = breadcrumbs_ul.find_all('li', class_='breadcrumb') if breadcrumbs_ul else []
    breadcrumb_list = [li.find('span').text.strip() for li in breadcrumbs_li_list]

    # Create a dictionary to store the data
    data_dict = {
        "url": url,
        "title": title,
        "breadcrumbs": breadcrumb_list,
        "price": price,
        "image": image_src,
        "mpn": mfg
        
    }

    # Append the dictionary to the data list
    data_list.append(data_dict)

# Create a DataFrame from the data list
result_df = pd.DataFrame(data_list)

# Write the DataFrame to a CSV file
result_df.to_csv('skygeek_data.csv', index=False)
```
